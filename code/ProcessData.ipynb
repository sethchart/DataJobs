```python
from JobsDb import JobsDb
from DataProcessor import data_processor
from tqdm import tqdm
import pickle
```

```python
db = JobsDb()
df = db.load_table_as_df('jobs')
db.close()
df = df.iloc[9680:]
data = df.copy()
data = data.reset_index().drop(['id', 'index'], axis=1)
descriptions = data['description']
titles = data['title']
```

```python
processed_titles = data_processor(tqdm(titles))
```

```python
processed_titles
```

```python
processed_descriptions = data_processor(tqdm(descriptions))
```

```python
data_records = data.to_dict('records')
for post, description_tokens, title_tokens in tqdm(zip(data_records, processed_descriptions, processed_titles)):
    post['description_tokens'] = description_tokens
    post['title_tokens'] = title_tokens
```

```python
data_records[1]
```

```python
with open('../data/processed_data.pkl', mode='wb') as file:
    pickle.dump(data_records, file)
```
